```python
#Statistics-advance-1-assignment-3 
#Question.1 What is Estimation Statistics? Explain point estimate and interval estimate.
#Answer.1 Estimation in statistics refers to the process of using sample data to make educated 
#guesses or estimates about population parameters. Estimation is a key component of inferential 
#statistics, where we aim to draw conclusions about a population based on information obtained
#from a sample.

#There are two main types of estimation in statistics: point estimation and interval estimation.

#Point Estimate:
#A point estimate is a single value that serves as the best guess or estimate of an unknown population
#parameter based on sample data. It provides a specific numerical value that is used as a "point" on 
#the parameter scale.

#For example, if we want to estimate the population mean based on a sample mean , the 
#sample mean would be the point estimate of the population mean. Similarly, the sample proportion 
#could serve as a point estimate of the population proportion 


#Point estimates are simple and easy to understand, but they have limitations. They don't provide 
#any information about the uncertainty or variability associated with the estimate. Additionally,
#different samples from the same population can result in different point estimates.

#Interval Estimate (Confidence Interval):
#An interval estimate, also known as a confidence interval, provides a range of values within which 
#the true population parameter is likely to fall with a certain level of confidence. It takes
#into account the variability of the sample data and provides a more informative estimate 
#compared to a single point estimate.

#Confidence intervals are typically expressed in the form: estimate ± margin of error. The margin of 
#error depends on the level of confidence chosen (e.g., 95%, 99%) and the variability of the sample data.

#Interval estimates provide a measure of uncertainty and allow us to quantify the range of
#likely values for the population parameter. They also take into account the variability 
#inherent in sample data.

#In summary, estimation in statistics involves making educated guesses about population
#parameters based on sample data. Point estimates provide single numerical values as 
#estimates, while interval estimates (confidence intervals) provide a range of values 
#within which the true parameter is likely to fall with a specified level of confidence. 
#Interval estimates offer more information about the uncertainty associated with the 
#estimate compared to point estimates alone.
```

```python
#Question.2 Write a Python function to estimate the population mean using a sample mean and standard
#deviation.
#Answer.2 : 
import scipy.stats as stats

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    """
    Estimate the population mean using a confidence interval.

    Parameters:
    sample_mean (float): The sample mean.
    sample_std (float): The sample standard deviation.
    sample_size (int): The sample size.
    confidence_level (float, optional): The desired confidence level (default is 0.95).

    Returns:
    tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """
    critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = critical_value * (sample_std / (sample_size ** 0.5))
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example usage:
sample_mean = 75.5
sample_std = 12.2
sample_size = 100
confidence_level = 0.95

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level)
print(f"95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
#Question.3 What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
#Answer.3 : Hypothesis testing is a fundamental concept in statistics that involves making
#decisions about a population based on sample data. It is a formal process used to assess whether
#a statement or assumption about a population parameter is likely to be true or not, given the
#available evidence from a sample.

#In hypothesis testing, there are two competing statements:

#Null Hypothesis (H0): This is the default assumption or statement that there is no effect,
#no difference, or no relationship in the population. It represents the status quo.
#Alternative Hypothesis (Ha or H1): This is the statement that contradicts the null hypothesis 
#and suggests that there is an effect, a difference, or a relationship in the population.
#The goal of hypothesis testing is to evaluate the evidence from the sample and determine 
#whether it provides enough support to reject the null hypothesis in favor of the alternative
#hypothesis.

#Steps in Hypothesis Testing:

#Formulate the null and alternative hypotheses.
#Collect and analyze sample data.
#Calculate a test statistic that measures the difference between the sample data and what is 
#expected under the null hypothesis.
#Determine the p-value, which represents the probability of observing a test statistic as
#extreme as, or more extreme than, the one calculated from the sample data, assuming that 
#the null hypothesis is true.
#Compare the p-value to a predetermined significance level (alpha) to make a decision:
#If p-value ≤ alpha: Reject the null hypothesis.
#If p-value > alpha: Fail to reject the null hypothesis.

#Importance of Hypothesis Testing:

#Informed Decision-Making: Hypothesis testing provides a systematic framework for making informed 
#decisions based on data. It helps answer questions and validate assumptions, which is crucial in
#various fields, including science, medicine, business, and social sciences.

#Scientific Research: Hypothesis testing is a fundamental tool in scientific research. Researchers
#use it to evaluate theories, test hypotheses, and draw conclusions about the underlying population.

#Quality Control: Hypothesis testing is used in quality control to determine whether a process or 
#product meets certain specifications or standards.

#Policy Making: Hypothesis testing provides evidence that can inform policy decisions. For example,
#in public health, it can help determine the effectiveness of interventions.

#Statistical Inference: Hypothesis testing allows us to generalize results from a sample to a
#larger population. It helps estimate population parameters and assess the reliability of 
#those estimates.

#Reducing Bias: Hypothesis testing provides an objective way to evaluate claims and reduce
#bias or subjectivity in decision-making.

#Problem Solving: In business and industry, hypothesis testing helps identify problems
#and find solutions by testing the effectiveness of different strategies.

#In summary, hypothesis testing is a powerful tool that helps us make objective and evidence-based
#decisions about population parameters and relationships. It plays a crucial role in scientific 
#inquiry, research, quality assurance, policy making, and problem-solving in various domains.
```

```python
#Question.5 Write a Python script to conduct a hypothesis test on the difference between two population
#means,given a sample from each population.
#Answer.5 :
import numpy as np
from scipy.stats import ttest_ind

# Generate example sample data
np.random.seed(0)
sample_size = 30
mean_male = 175  # Mean weight of male college students
std_dev_male = 10
sample_male = np.random.normal(mean_male, std_dev_male, sample_size)

mean_female = 160  # Mean weight of female college students
std_dev_female = 12
sample_female = np.random.normal(mean_female, std_dev_female, sample_size)

# Perform independent samples t-test
t_statistic, p_value = ttest_ind(sample_male, sample_female)

# Define significance level (alpha)
alpha = 0.05

# Compare p-value to significance level to make a decision
if p_value < alpha:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

# Print results
print("Results of Hypothesis Test")
print("--------------------------")
print("Sample Size:", sample_size)
print("Mean (Male):", np.mean(sample_male))
print("Mean (Female):", np.mean(sample_female))
print("t-Statistic:", t_statistic)
print("p-Value:", p_value)
print("Conclusion:", conclusion)
```

```python
#Queston.8 Define p-value and explain its significance in hypothesis testing.
#Answer.8 The p-value, short for "probability value," is a fundamental concept in hypothesis testing.
#It represents the probability of obtaining a test statistic as extreme as, or more extreme than, the
#one calculated from the sample data, assuming that the null hypothesis is true. In other words, 
#the p-value quantifies the evidence against the null hypothesis provided by the sample data.

#Significance of the p-value in Hypothesis Testing:

#The p-value serves as a crucial tool in hypothesis testing for making decisions about whether to
#reject or fail to reject the null hypothesis. Here's how the p-value is used and its significance:

#Decision Rule:

#If the p-value is small (typically smaller than the chosen significance level α), it suggests that
#the observed sample data is unlikely under the null hypothesis. This leads to the rejection of the
#null hypothesis in favor of the alternative hypothesis.
#If the p-value is large, it suggests that the observed sample data is reasonably likely
#under the null hypothesis. This leads to the failure to reject the null hypothesis.

#Comparing to the Significance Level (α):

#If p-value ≤ α: Reject the null hypothesis.
#If p-value > α: Fail to reject the null hypothesis.

#Interpretation:

#A small p-value suggests that the observed sample data provides strong evidence against the null
#hypothesis. It indicates that the observed effect or difference is unlikely due to random chance alone.
#A large p-value suggests that the observed sample data does not provide strong evidence against the 
#null hypothesis. It indicates that the observed effect or difference could plausibly occur due to
#random variability.

#Importance of the p-value:

#The p-value provides a standardized measure of the strength of evidence against the null hypothesis. 
#It helps researchers objectively assess the results of a hypothesis test.
#The p-value allows researchers to make informed decisions based on statistical principles rather than 
#subjective judgment.
#The p-value helps ensure that conclusions drawn from hypothesis tests are transparent, reproducible,
#and based on quantitative evidence.

#Note on Interpretation:

#The p-value does not directly provide information about the magnitude or practical importance of an
#effect. It only assesses the evidence against the null hypothesis.
#A small p-value does not prove that the alternative hypothesis is true; it suggests that the null
#hypothesis is unlikely to be true given the sample data.
```

```python
#Question.9 Generate a Student's t-distribution plot using Python's matplotlib library, with the
#degrees of freedom parameter set to 10.
#Answer.9 : 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Generate x values (range of t-distribution)
x = np.linspace(-4, 4, 500)

# Calculate the corresponding y values (probability density function)
y = stats.t.pdf(x, df)

# Plot the t-distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"t-Distribution (df = {df})")
plt.title("Student's t-Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Question.10 Write a Python program to calculate the two-sample t-test for
#independent samples, given two random samples of equal size and a null hypothesis 
#that the population means are equal.
#Answer.10 : 
import numpy as np
import scipy.stats as stats

# Generate two random samples of equal size
np.random.seed(0)
sample_size = 30
sample1 = np.random.normal(50, 10, sample_size)
sample2 = np.random.normal(55, 10, sample_size)

# Perform two-sample t-test assuming equal variances (pooled t-test)
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Define significance level (alpha)
alpha = 0.05

# Compare p-value to significance level to make a decision
if p_value < alpha:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

# Print results
print("Results of Two-Sample T-Test")
print("---------------------------")
print("Sample Size:", sample_size)
print("Sample 1 Mean:", np.mean(sample1))
print("Sample 2 Mean:", np.mean(sample2))
print("t-Statistic:", t_statistic)
print("p-Value:", p_value)
print("Conclusion:", conclusion)
```

```python
#Question.11 What is Student’s t distribution? When to use the t-Distribution.
#Answer.11 The Student's t-distribution, often simply referred to as the t-distribution, is 
#a probability distribution that is used in statistics to perform hypothesis tests and make 
#inferences about population parameters, particularly when the sample size is small or the 
#population standard deviation is unknown. It is a fundamental distribution in statistical 
#analysis and is widely used in various fields for hypothesis testing, confidence interval
#estimation, and other inferential procedures.

#Characteristics of the t-Distribution:

#The t-distribution is symmetric and bell-shaped, similar in appearance to the normal distribution,
#but with heavier tails.
#The shape of the t-distribution depends on the degrees of freedom (df). As the degrees of freedom
#increase, the t-distribution approaches the standard normal 
#distribution (Z-distribution).

#When to Use the t-Distribution:

#The t-distribution is used under the following circumstances:

#Small Sample Size: When dealing with small sample sizes (typically n<30) where the population standard 
#deviation is unknown, the t-distribution is used to 
#make inferences about the population mean. In such cases, the t-distribution provides
#more accurate results than the normal distribution.

#Population Standard Deviation Unknown: When the population standard deviation is unknown 
#and needs to be estimated from the sample data, the t-distribution is used to account for
#the additional uncertainty introduced by the estimation.

#Hypothesis Testing and Confidence Intervals: The t-distribution is commonly used in hypothesis
#testing to assess whether a sample mean is significantly different from a hypothesized population 
#mean. It is also used to construct confidence intervals for population parameters when the population
#standard deviation is unknown.

#Comparing Means: When comparing means of two independent samples or conducting paired-sample 
#comparisons (dependent samples), the t-distribution is used to assess the significance of 
#observed differences.

#Quality Control and Process Improvement: The t-distribution is used in quality control to
#determine whether a process is operating within specified limits or to assess the effectiveness 
#of process improvements.

#Experimental Design: When conducting experiments and analyzing data, the t-distribution is used 
#to make inferences about treatment effects or differences between groups.

#Real-Life Data Analysis: The t-distribution is applied in various real-life scenarios, such as
#medical research, social sciences, economics, and engineering, where sample sizes are often 
#limited and assumptions of normality and equal variances are not always met.

#In summary, the t-distribution is a vital tool in statistics, particularly when dealing with
#small sample sizes and situations where the population standard deviation is unknown. It 
#provides a more accurate and robust framework for making statistical inferences and conducting
#hypothesis tests in such cases.
```

```python
#Question.13 A coffee shop owner wants to estimate the average daily revenue for their shop. 
#They take a random
#sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
#Estimate the population mean revenue with a 95% confidence interval.
#Answer.13 : 
import numpy as np
import scipy.stats as stats

# Given sample information
sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

# Calculate the critical value (z-score) for a 95% confidence interval
critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

# Print the confidence interval
print("95% Confidence Interval for Population Mean Revenue:")
print(f"Lower Bound: ${confidence_interval_lower:.2f}")
print(f"Upper Bound: ${confidence_interval_upper:.2f}")
```

```python
#Question.14 A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
#clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
#standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
#Answer.14 : 
import numpy as np
import scipy.stats as stats

# Given sample information
sample_mean = 8
sample_std_dev = 3
sample_size = 100
hypothesized_mean = 10
alpha = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_std_dev / np.sqrt(sample_size))

# Degrees of freedom for a one-sample t-test
degrees_of_freedom = sample_size - 1

# Calculate the critical value from the t-distribution
critical_value = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

# Perform the hypothesis test
if np.abs(t_statistic) > critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

# Print results
print("Results of Hypothesis Test")
print("--------------------------")
print("Sample Size:", sample_size)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std_dev)
print("Hypothesized Population Mean:", hypothesized_mean)
print("t-Statistic:", t_statistic)
print("Degrees of Freedom:", degrees_of_freedom)
print("Critical Value:", critical_value)
print("Conclusion:", conclusion)
```

```python
#Question.15 An electronics company produces a certain type of product with a mean weight of 5 pounds and a
#standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
#is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
#pounds with a significance level of 0.01.
#Answer.15 : 
import numpy as np
import scipy.stats as stats

# Given sample information
sample_mean = 4.8
population_std_dev = 0.5
sample_size = 25
hypothesized_mean = 5
alpha = 0.01

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (population_std_dev / np.sqrt(sample_size))

# Degrees of freedom for a one-sample t-test
degrees_of_freedom = sample_size - 1

# Calculate the critical value from the t-distribution
critical_value = stats.t.ppf(alpha, df=degrees_of_freedom)

# Perform the hypothesis test
if t_statistic < critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

# Print results
print("Results of Hypothesis Test")
print("--------------------------")
print("Sample Size:", sample_size)
print("Sample Mean:", sample_mean)
print("Population Standard Deviation:", population_std_dev)
print("Hypothesized Population Mean:", hypothesized_mean)
print("t-Statistic:", t_statistic)
print("Degrees of Freedom:", degrees_of_freedom)
print("Critical Value:", critical_value)
print("Conclusion:", conclusion)
```

```python
#Question.16 Two groups of students are given different study materials to prepare for a test. The first group (n1 =
#30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
#score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
#groups are equal with a significance level of 0.01.
#Answer.16 : 
import numpy as np
import scipy.stats as stats

# Given sample information
sample1_mean = 80
sample1_std_dev = 10
sample1_size = 30

sample2_mean = 75
sample2_std_dev = 8
sample2_size = 40

alpha = 0.01

# Calculate the pooled standard error
pooled_std_error = np.sqrt((sample1_std_dev ** 2 / sample1_size) + (sample2_std_dev ** 2 / sample2_size))

# Calculate the t-statistic
t_statistic = (sample1_mean - sample2_mean) / pooled_std_error

# Degrees of freedom for a two-sample t-test
degrees_of_freedom = sample1_size + sample2_size - 2

# Calculate the critical value from the t-distribution
critical_value = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

# Perform the hypothesis test
if np.abs(t_statistic) > critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

# Print results
print("Results of Hypothesis Test")
print("--------------------------")
print("Sample 1 Size:", sample1_size)
print("Sample 1 Mean:", sample1_mean)
print("Sample 1 Standard Deviation:", sample1_std_dev)

print("Sample 2 Size:", sample2_size)
print("Sample 2 Mean:", sample2_mean)
print("Sample 2 Standard Deviation:", sample2_std_dev)

print("Degrees of Freedom:", degrees_of_freedom)
print("t-Statistic:", t_statistic)
print("Critical Value:", critical_value)
print("Conclusion:", conclusion)
```
