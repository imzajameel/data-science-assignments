```python
#Flask Assignment.1 
#Question.1 : What is Flask Framework? What are the advantages of Flask Framework?
#Answer.1 : Flask is a micro web framework for Python, designed to make it easy to build web applications
#quickly and with minimal boilerplate code. It is classified as a micro framework because it provides only 
#the essential tools for building a web application, leaving developers with the flexibility to choose additional 
#libraries and tools based on their specific project requirements. Flask follows the WSGI (Web Server Gateway Interface)
#standard and can be used with various web servers.

#Advantages of Flask Framework:

#Minimalistic and Lightweight: Flask is designed to be simple and lightweight, making it easy for developers to
#understand and work with. It doesn't impose a lot of structure on your application, giving you the freedom to 
#organize your code the way you want.

#Flexibility: Being a micro framework, Flask provides just the basic tools for building a web application.
#This allows you to choose the components you need and integrate them seamlessly into your project, giving 
#you greater control over the development process.

#Easy to Get Started: Flask's simple and intuitive API makes it easy for both beginners and experienced
#developers to get started quickly. You can create a basic web application with just a few lines of code.

#Extensible: While Flask itself is minimal, it is highly extensible. You can easily add various extensions
#to your Flask application to add functionality such as database integration, authentication, caching, and more.

#Wide Range of Libraries: Flask integrates well with other Python libraries and tools. This means you can leverage 
#the extensive Python ecosystem for various tasks, such as data manipulation, machine learning, and more.

#Built-in Development Server: Flask comes with a built-in development server, which is convenient for testing and 
#debugging your application during development without needing to configure a separate web server.

#Great for Small to Medium-sized Applications: Flask is particularly well-suited for building small to medium-sized
#web applications, RESTful APIs, and prototypes where a full-fledged framework might be overkill.

#Community and Documentation: Flask has a strong and supportive community that provides a wealth of resources,
#including documentation, tutorials, and third-party extensions. This makes it easier to find solutions to
#common problems and learn how to use the framework effectively.

#Jinja2 Templating: Flask uses the Jinja2 template engine, which enables you to separate the presentation layer
#from the business logic of your application.

#URL Routing: Flask includes a powerful URL routing mechanism that allows you to map URLs to specific view functions,
#making it easy to organize your application's routes.

#It's important to note that while Flask offers numerous advantages, it might not be the best fit for every project.
#For larger applications with complex requirements, a more feature-rich framework like Django might be a better choice. 
#Flask's strength lies in its simplicity, flexibility, and ability to rapidly develop smaller-scale applications.
```

```python
#Question.2 : Create a simple Flask application to display ‘Hello World!!’.
#Answer.2 : 
from flask import Flask

# Create a Flask application instance
app = Flask(__name__)

# Define a route for the root URL ("/")
@app.route("/")
def hello_world():
    return "Hello World!!"

# Run the application if this script is executed
if __name__ == "__main__":
    app.run()
```

```python
#Question.3 : What is App routing in Flask? Why do we use app routes?
#Answer.3 : App routing in Flask refers to the process of defining URL routes for different parts
#of your web application. Routes determine how the application responds to different URLs that users
#enter in their browsers. In Flask, you use the @app.route() decorator to associate a URL with a Python
#function. When a user accesses a specific URL, Flask maps that URL to the corresponding function, which
#generates the appropriate response.

#Reasons why we use app routes in Flask : 

#URL Structure: App routes allow you to define the structure of your application's URLs. You can create
#clean and meaningful URLs that reflect the different sections and functionalities of your application.

#Organization: By defining routes, you organize your application's functionality into distinct views.
#This separation of concerns makes your codebase more maintainable and easier to navigate.

#Modular Development: Flask routes encourage modular development. You can create separate view functions 
#for different parts of your application, making it easier to collaborate with other developers and manage changes.

#Dynamic Content: Routes can include placeholders for dynamic content in the URL. For instance, you can
#define a route like /user/<username> to handle profiles for different users dynamically.

#RESTful APIs: If you're building a RESTful API, app routes help define the endpoints and actions available to 
#clients.

#HTTP Methods: Flask routes can be associated with specific HTTP methods (GET, POST, etc.). This enables you 
#to handle different types of requests differently, allowing you to create more complex interactions.

#Readable Code: Using app routes with meaningful function names makes your code more readable. 
#Other developers can easily understand which functions correspond to specific URL paths.

#Flexibility: Routes allow you to map URLs to arbitrary Python functions, giving you the flexibility 
#to generate responses based on your application's logic.
```

```python
#Question.4 : Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
#route to show the following details:
#Company Name: ABC Corporation
#Location: India
#Contact Detail: 999-999-9999
#Answer.4 : 
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

@app.route("/")
def company_details():
    details = """
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    """
    return details

if __name__ == "__main__":
    app.run()
```

```python
#Question.5 : What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
#url_for() function.
#Answer.5 : In Flask, the url_for() function is used for URL building. It generates a URL for a given 
#endpoint (view function) and can also handle dynamic parts of the URL, such as placeholders for route 
#parameters. This function is useful because it abstracts the URL generation process and makes your code 
#more flexible to changes in URL structure.

#Here's a Python code example that demonstrates the usage of the url_for() function :
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

@app.route("/user/<username>")
def user_profile(username):
    return f"Profile page of {username}"

if __name__ == "__main__":
    with app.test_request_context():
        print("Generated URLs using url_for():")
        print(url_for("welcome"))  # Generates URL for the welcome() function
        print(url_for("user_profile", username="john"))  # Generates URL for user_profile() with username parameter
```
