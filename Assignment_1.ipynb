```python
#Question.5 : Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution
#of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to
#these features to improve normality?
#Answer.5 : Exploratory Data Analysis (EDA) is a crucial step in understanding the characteristics and 
#distributions of features in a dataset. In this context, we'll work with the wine quality dataset to perform 
#EDA and identify features that exhibit non-normality. We'll also discuss possible transformations to improve normality.

#Performing EDA:

#Start by loading the wine quality dataset and examining its structure, summary statistics, and data distributions.
#Create visualizations like histograms, box plots, and Q-Q plots for each feature to assess their distributions.
#Evaluate the mean, median, and skewness of the feature distributions to identify non-normality.

#Identifying Non-Normal Features:

#Features that exhibit non-normality often have skewed or heavy-tailed distributions. Common signs of non-normality 
#include a significant skewness value and deviations from a symmetric shape in histograms and Q-Q plots.

#Possible Transformations for Non-Normal Features:

#Several transformations can be applied to improve the normality of non-normal features:
    
#Logarithmic Transformation: Use a logarithmic transformation (e.g., natural logarithm) to reduce right-skewness.
#Box-Cox Transformation: Apply the Box-Cox transformation to make the data more normally distributed. The transformation
#parameter is estimated to find the best fit.
#Square Root Transformation: Take the square root of the data to reduce right-skewness.
#Reciprocal Transformation: Use the reciprocal (1/x) transformation to mitigate left-skewness.
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the wine quality dataset (replace 'winequality.csv' with your dataset file)
data = pd.read_csv('red_wine.csv', sep=';')

# Get the numeric columns
numeric_cols = data.select_dtypes(include=['number'])

# Set the figure size for better visualization
plt.figure(figsize=(15, 12))

# Create a list to store non-normal features
non_normal_features = []

# Create subplots for each numeric feature
for i, feature in enumerate(numeric_cols.columns, 1):
    plt.subplot(3, 4, i)
    
    # Create a histogram to visualize the distribution
    plt.hist(numeric_cols[feature], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    
    # Calculate skewness
    skewness = stats.skew(numeric_cols[feature])
    plt.title(f'Skewness: {skewness:.2f}', loc='left', fontsize=10)
    
    # Define a threshold for skewness (e.g., 0.5) to identify non-normal features
    if abs(skewness) > 0.5:
        non_normal_features.append(feature)

# Adjust subplots layout
plt.tight_layout()

# Show the plots
plt.show()

# Identify non-normal features
if non_normal_features:
    print("Features that exhibit non-normality:")
    for feature in non_normal_features:
        print(f"- {feature}")
    print("\nPotential transformations to improve normality:")
    print("- Logarithmic Transformation")
    print("- Box-Cox Transformation")
    print("- Square Root Transformation")
    print("- Reciprocal Transformation")
else:
    print("No features exhibit significant non-normality.")

```

```python
#Question.6 : Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
#features. What is the minimum number of principal components required to explain 90% of the variance in
#the data?
#Answer.6 : 
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import requests
from io import StringIO

# Define the URL for the red wine dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

# Load the dataset from the URL
response = requests.get(url)
data = pd.read_csv(StringIO(response.text), sep=';')

# Separate the target variable (quality) from the features
X = data.drop('quality', axis=1)

# Standardize the features (mean=0, std=1)
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Create a PCA model
pca = PCA()

# Fit the PCA model to the standardized data
pca.fit(X_std)

# Calculate the explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate the cumulative explained variance
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Find the minimum number of components to explain at least 90% of the variance
num_components_90_variance = (cumulative_explained_variance >= 0.90).sum()

print(f"Minimum number of principal components to explain 90% of the variance: {num_components_90_variance}")
```
