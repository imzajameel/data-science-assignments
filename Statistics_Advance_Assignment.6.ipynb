```python
#Statistics_Advance_Assignment_6 
#Question.1 Explain the assumptions required to use ANOVA and provide examples of violations
#that could impact the validity of the results.
#Answer.1 Analysis of Variance (ANOVA) is a statistical method used to compare means between three or
#more groups. It is an extension of the t-test, which is used to compare means between two groups.
#ANOVA allows you to determine if there are significant differences among the means of multiple groups. 
#However, ANOVA has certain assumptions that need to be met for its results to be valid. Violations of 
#these assumptions can impact the reliability and interpretation of ANOVA results.

#Assumptions of ANOVA:

#Independence: The observations in each group should be independent of each other. This means that the 
#value of one observation should not be influenced by the value of another observation within the same group.

#Normality: The data within each group should follow a normal distribution. This assumption is especially 
#important when sample sizes are small. If the data is heavily skewed or contains outliers, the normality
#assumption may be violated.

#Homogeneity of Variance (Homoscedasticity): The variances of the groups should be approximately equal. 
#This assumption ensures that the variability within each group is consistent across all groups.

#Examples of Violations and Impact on ANOVA:

#Non-Independence: If observations are not independent, such as when data is collected from repeated measurements on 
#the same subjects, it can lead to correlated errors and inflated Type I error rates.

#Non-Normality: If the data within each group is not normally distributed, ANOVA results may be invalid. This can 
#lead to inaccurate p-values and confidence intervals. If the data is highly skewed, non-parametric alternatives 
#like the Kruskal-Wallis test may be more appropriate.

#Heteroscedasticity: If the variances of the groups are not equal, ANOVA may be less robust. In such cases, Welch's 
#ANOVA or transformations of the data may be considered.

#Outliers: Outliers can affect both normality and homogeneity of variance assumptions. Extreme values may 
#disproportionately affect group means, leading to incorrect conclusions about group differences.

#Unequal Sample Sizes: ANOVA assumes that sample sizes across groups are approximately equal. If one group 
#has a significantly larger or smaller sample size than others, it can impact the power of the test and affect 
#the validity of results.

#Interaction Effects: If there are interaction effects between factors (e.g., when the effect of one factor depends
#on the level of another factor), the assumptions of ANOVA may be violated. In such cases, more complex analyses like
#factorial ANOVA or regression models may be necessary.

#Violation of Study Design: ANOVA assumes that groups are formed randomly or by a well-defined experimental design. If 
#groups are not properly randomized or assigned, the validity of ANOVA results can be compromised.

#When conducting ANOVA, it's important to assess whether these assumptions are met, and if not, consider appropriate
#alternatives or transformations of the data. If assumptions are significantly violated, it's advisable to consult 
#with a statistician or use alternative methods to analyze the data.
```

```python
#Question.2 What are the three types of ANOVA, and in what situations would each be used?
#Answer.2 There are three main types of Analysis of Variance (ANOVA) techniques, each designed 
#to address specific research questions and situations. The three types of ANOVA are:

#One-Way ANOVA: One-Way ANOVA is used when you have one independent variable (factor) with three
#or more levels (groups) and you want to determine if there are any statistically significant 
#differences among the means of these groups. It is used to compare means across multiple groups simultaneously.

#Example: A researcher wants to determine if there is a significant difference in the average test 
#scores among students from three different schools (School A, School B, and School C).

#Two-Way ANOVA: Two-Way ANOVA is used when you have two independent variables (factors) and you want 
#to examine the effects of these factors on a dependent variable. It helps you understand whether the
#two factors have independent or interactive effects on the dependent variable.

#Two-Way ANOVA without Interaction: Used when the two factors do not interact and you are interested in 
#the main effects of each factor.

#Example: A study investigates the effects of two different diets (Diet A and Diet B) and two different 
#exercise regimes (Exercise X and Exercise Y) on weight loss.

#Two-Way ANOVA with Interaction: Used when the two factors interact and you want to examine both the main
#effects of each factor and their interaction.

#Example: A study examines how student performance (dependent variable) is influenced by two factors:
#teaching method (Method A, Method B) and class size (Small, Large).

#Repeated Measures ANOVA (Within-Subjects ANOVA): Repeated Measures ANOVA is used when you have one
#group of participants measured at multiple time points or under different conditions. It's used to
#analyze data in which each participant is exposed to all levels of the independent variable.

#Example: A study measures participants' heart rates before and after performing different exercise 
#routines (Condition A, Condition B, Condition C) to determine if there are significant differences.

#In summary:

#Use One-Way ANOVA when you have one factor with multiple levels and want to compare means among groups.
#Use Two-Way ANOVA when you have two factors and want to examine their effects on a dependent variable, 
#with or without interaction.
#Use Repeated Measures ANOVA when you have one group measured repeatedly or under different conditions to
#assess changes across time or conditions.
#The choice of ANOVA type depends on the experimental design and research questions you are addressing. 
#It's important to choose the appropriate type of ANOVA to ensure that your analysis aligns with your 
#study design and objectives.
```

```python
#Question.3 What is the partitioning of variance in ANOVA, and why is it important to 
#understand this concept?
#Answer.3 Partitioning of variance is a fundamental concept in Analysis of Variance (ANOVA) 
#that involves breaking down the total variation in a dataset into different components attributed 
#to different sources or factors. It helps us understand how much of the total variability in the 
#data can be explained by the factors being studied and how much is due to random variability or error.
#This partitioning allows us to quantify the contributions of various factors to the observed differences
#and helps in making informed conclusions about the significance of those factors.

#In ANOVA, the total variability in the data is divided into three main components:

#Between-Group Variability (Treatment Variability): This component represents the variation in the data that 
#is due to the differences between the groups or levels of the independent variable (factor). It reflects the 
#effect of the factor being studied.

#Within-Group Variability (Error Variability): This component represents the variation within each group that 
#is not explained by the factor being studied. It includes random variability, measurement error, and any other
#unaccounted sources of variation.

#Total Variability: The total variability in the data, also known as the total sum of squares (SST), is the sum of 
#the between-group variability and the within-group variability. It represents the overall variation in the data 
#without considering any specific factors.

#Mathematically, the partitioning of variance can be expressed as:

#SSTotal = SSBetween + SSWithin

#Understanding Partitioning of Variance Important because : 

#Quantifying Factor Effects: Partitioning of variance allows us to quantify the extent to which the differences 
#among groups can be attributed to the factor being studied. It helps answer questions such as, "How much
#of the variation in test scores is due to different teaching methods?"

#Hypothesis Testing and Significance: By comparing the ratio of between-group variability to within-group 
#variability, ANOVA determines whether the observed differences among groups are statistically significant
#or if they could have occurred by random chance.

#Interpreting Results: Understanding partitioning of variance helps researchers interpret ANOVA results. A large
#portion of between-group variability relative to within-group variability suggests a strong effect of the factor
#being studied, while a small portion indicates weaker influence.

#Study Design and Control: Partitioning of variance informs study design and allows researchers to control for 
#factors that contribute to variability. It aids in designing experiments that minimize noise and maximize the
#signal of interest.

#Comparing Different Models: Partitioning of variance is used to compare different models or statistical 
#techniques, helping researchers choose the best-fitting model for their data.

#In summary, partitioning of variance is a foundational concept in ANOVA that helps researchers understand 
#the sources of variability in their data, evaluate the significance of factors, make informed conclusions,
#and design effective experiments. It plays a crucial role in hypothesis testing, model comparison, and drawing 
#meaningful insights from data analysis.
```

```python
#Question.4 How would you calculate the total sum of squares (SST), explained sum of squares (SSE), and residual
#sum of squares (SSR) in a one-way ANOVA using Python?
#Answer.4 : 
import numpy as np
import scipy.stats as stats

# Example data for three groups
group1 = np.array([10, 12, 15, 18, 20])
group2 = np.array([25, 30, 32, 35, 40])
group3 = np.array([50, 55, 58, 60, 65])

# Combine the data into a single array
data = np.concatenate((group1, group2, group3))

# Calculate the overall mean
overall_mean = np.mean(data)

# Calculate the total sum of squares (SST)
sst = np.sum((data - overall_mean)**2)

# Calculate the group means
group1_mean = np.mean(group1)
group2_mean = np.mean(group2)
group3_mean = np.mean(group3)

# Calculate the explained sum of squares (SSE)
sse = np.sum((group1_mean - overall_mean)**2) * len(group1) + \
      np.sum((group2_mean - overall_mean)**2) * len(group2) + \
      np.sum((group3_mean - overall_mean)**2) * len(group3)

# Calculate the residual sum of squares (SSR)
ssr = sst - sse

# Print the results
print("Total Sum of Squares (SST):", sst)
print("Explained Sum of Squares (SSE):", sse)
print("Residual Sum of Squares (SSR):", ssr)
```

```python
#Question.5 In a two-way ANOVA, how would you calculate the main effects and interaction effects using Python?
#Answer.5 : 
import numpy as np
import scipy.stats as stats

# Example data for a two-way ANOVA
# Assuming two factors: Factor A (levels: A1, A2) and Factor B (levels: B1, B2, B3)
data = np.array([
    [10, 12, 15],  # Factor A: A1, Factor B: B1
    [18, 20, 22],  # Factor A: A1, Factor B: B2
    [25, 28, 30],  # Factor A: A1, Factor B: B3
    [30, 32, 35],  # Factor A: A2, Factor B: B1
    [38, 40, 42],  # Factor A: A2, Factor B: B2
    [45, 48, 50],  # Factor A: A2, Factor B: B3
])

# Calculate the overall mean
overall_mean = np.mean(data)

# Calculate the main effects
main_effect_A = np.sum((np.mean(data, axis=1) - overall_mean)**2) * data.shape[1]
main_effect_B = np.sum((np.mean(data, axis=0) - overall_mean)**2) * data.shape[0]

# Calculate the interaction effect
interaction_effect = np.sum((data - np.mean(data, axis=0) - np.mean(data, axis=1)[:, np.newaxis] + overall_mean)**2)

# Calculate the residual sum of squares (SSR)
sst = np.sum((data - overall_mean)**2)
ssr = sst - main_effect_A - main_effect_B - interaction_effect

# Print the results
print("Main Effect of Factor A:", main_effect_A)
print("Main Effect of Factor B:", main_effect_B)
print("Interaction Effect:", interaction_effect)
print("Residual Sum of Squares (SSR):", ssr)
```

```python
#Question.6 Suppose you conducted a one-way ANOVA and obtained an F-statistic of 5.23 and a p-value of 0.02.
#What can you conclude about the differences between the groups, and how would you interpret these
#results?
#Answer.6 : In a one-way ANOVA, the F-statistic is used to test the null hypothesis that there are no 
#significant differences between the means of the groups. The p-value associated with the F-statistic 
#indicates the probability of obtaining the observed F-statistic (or a more extreme value) if the null 
#hypothesis is true. A small p-value suggests that the observed differences between the groups are 
#statistically significant, leading to the rejection of the null hypothesis.

#Given your results:

#F-statistic: 5.23
#p-value: 0.02
#Interpretation:

#F-Statistic: The F-statistic value of 5.23 indicates the ratio of the variability between the group means
#to the variability within the groups. It tells us whether the observed differences between the group means 
#are larger than what we would expect due to random chance.

#p-value: The p-value of 0.02 is the probability of obtaining an F-statistic as extreme as 5.23 or more extreme,
#assuming that there are no real differences between the groups. Since the p-value is below the typical significance 
#level (such as 0.05), it suggests that the observed differences are statistically significant.

#Conclusion:
    
#Based on the results, you can conclude that there are statistically significant differences between at least
#some of the groups. In other words, there is evidence to reject the null hypothesis that all group means are 
#equal. The small p-value indicates that the likelihood of obtaining the observed F-statistic under the assumption 
#of no real differences between the groups is very low.

#Keep in mind that a significant result in ANOVA only tells you that there are differences among the groups,
#but it doesn't specifically identify which groups are different from each other. For that, you might need to
#conduct post hoc tests (such as Tukey's HSD or Bonferroni corrections) to determine which specific group means
#differ significantly.

#Additionally, it's important to consider the context of your study and the practical significance of the differences.
#A statistically significant difference doesn't necessarily imply a meaningful or important difference in a real-world
#context.
```

```python
#Question.7 In a repeated measures ANOVA, how would you handle missing data, and what are the potential
#consequences of using different methods to handle missing data?
#Answer.7 Handling missing data in a repeated measures ANOVA is an important consideration to ensure
#the validity and reliability of your analysis. Missing data can arise due to various reasons, such as 
#participants dropping out of the study or incomplete data collection. How you handle missing data can 
#impact the results and conclusions of your analysis. Here are some common methods for handling missing 
#data in a repeated measures ANOVA, along with potential consequences of using different methods:

#Listwise Deletion (Complete Case Analysis):

#This method involves removing any participant with missing data on any of the variables, resulting in a 
#reduced sample size.
#Consequences: Listwise deletion can lead to biased results if missingness is related to the outcome
#variable or if the missing data pattern is not completely random. It may also result in reduced power
#and efficiency due to the smaller sample size.

#Pairwise Deletion:

#Pairwise deletion retains participants with complete data for each specific analysis, using available data for
#each specific pair of variables.
#Consequences: Similar to listwise deletion, pairwise deletion can lead to biased results and reduced power if
#missing data are not missing completely at random. It can also complicate interpretation due to varying sample
#sizes across different analyses.

#Mean Imputation:

#Missing values are replaced with the mean of the available data for that variable.
#Consequences: Mean imputation can artificially reduce variability and may lead to biased estimates of means, 
#standard errors, and statistical significance. It assumes that the missing values have the same mean as the 
#observed values, which may not be valid.

#Last Observation Carried Forward (LOCF):

#Missing values are replaced with the last observed value for that participant.
#Consequences: LOCF assumes that the participant's condition remains constant between time points, which may
#not be accurate. This method can lead to biased estimates, especially if there is a trend or change in the data.

#Linear Interpolation:

#Missing values are estimated based on linear interpolation between adjacent time points.
#Consequences: Linear interpolation assumes a linear relationship between time points, which may not hold true
#for all data. It can also introduce noise if the underlying assumption is violated.

#Multiple Imputation:

#Missing values are imputed multiple times to create multiple complete datasets, and the analyses are conducted on 
#each dataset. The results are then combined to provide more robust estimates.
#Consequences: Multiple imputation is a more sophisticated approach that can provide unbiased estimates if the
#assumptions of missingness are met. However, it requires more complex computations and may not be practical for
#small sample sizes.
#It's important to carefully consider the reasons for missing data, the assumptions underlying each method, and 
#the potential consequences of your chosen approach. No method is universally best, and the choice depends on the
#nature of the data and the specific research question. Sensitivity analyses and exploring different methods can 
#help assess the robustness of your results to missing data handling.
```

```python
#Question.8 What are some common post-hoc tests used after ANOVA, and when would you use each one? Provide
#an example of a situation where a post-hoc test might be necessary.
#Answer.8 After conducting an ANOVA and finding a significant overall effect, post-hoc tests are used to make 
#pairwise comparisons between specific groups to determine which groups are significantly different from each other.
#These tests help control the familywise error rate that can arise from conducting multiple pairwise comparisons. 
#Some common post-hoc tests include:

#Tukey's Honestly Significant Difference (HSD):

#Tukey's HSD is a widely used post-hoc test that compares all possible pairs of group means and calculates a 
#critical value for significance.
#Use Tukey's HSD when you have conducted a one-way ANOVA and want to compare all possible pairs of group means. 
#It controls the overall Type I error rate.

#Bonferroni Correction:

#The Bonferroni correction adjusts the significance level for each individual comparison to maintain an overall 
#desired error rate (usually 0.05).
#Use the Bonferroni correction when you have conducted multiple pairwise comparisons and want to control the 
#familywise error rate. It's conservative and helps reduce the chances of making a Type I error.

#Dunnett's Test:

#Dunnett's test is used when you have a control group and want to compare other groups to the control group while
#controlling the familywise error rate.
#Use Dunnett's test when you have a control group and want to focus on comparisons between other groups and the 
#control group.

#Scheffe's Test:

#Scheffe's test is a conservative post-hoc test that can be used for all possible pairwise comparisons, similar to
#Tukey's HSD.
#Use Scheffe's test when you want to control the overall familywise error rate for a large number of pairwise 
#comparisons. It is less sensitive to assumptions of equal variances and normality compared to Tukey's HSD.

#Games-Howell Test:

#The Games-Howell test is a non-parametric post-hoc test that doesn't assume equal variances among groups.
#Use the Games-Howell test when the assumption of equal variances is violated and you need to make pairwise
#comparisons between groups.
#Example Situation:

#Suppose you conducted a study to compare the effectiveness of four different diets (A, B, C, D) on weight loss.
#After performing a one-way ANOVA, you find a significant difference in mean weight loss among the diet groups.
#However, the ANOVA does not tell you which specific diets are significantly different from each other.

#To determine which diets are significantly different, you would conduct post-hoc tests. For instance, if you
#decide to use Tukey's HSD, you would run pairwise comparisons for all possible combinations of diet groups
#(A vs. B, A vs. C, A vs. D, B vs. C, B vs. D, C vs. D). Tukey's HSD would provide you with confidence intervals
#and p-values for each comparison, allowing you to identify which diets have significantly different effects on
#weight loss.

#Post-hoc tests help you make more specific and informative comparisons between groups after finding a significant 
#overall effect in your ANOVA.
```

```python
#Question.9 A researcher wants to compare the mean weight loss of three diets: A, B, and C. They collect data from
#50 participants who were randomly assigned to one of the diets. Conduct a one-way ANOVA using Python
#to determine if there are any significant differences between the mean weight loss of the three diets.
#Report the F-statistic and p-value, and interpret the results.
#Answer.9 : 
import numpy as np
import scipy.stats as stats

# Example data for weight loss (three diets: A, B, and C)
diet_A = np.array([2.5, 3.0, 4.0, 2.8, 3.5, 3.2, 2.6, 2.9, 3.8, 3.2,
                   2.7, 3.5, 3.0, 3.1, 2.9, 2.4, 2.8, 3.2, 3.6, 3.4,
                   2.7, 3.9, 3.1, 2.8, 3.3, 2.5, 3.2, 3.4, 2.9, 2.7,
                   2.8, 3.0, 2.6, 3.3, 3.5, 3.6, 2.8, 2.9, 3.1, 3.2,
                   2.6, 3.0, 2.7, 3.4, 3.2, 3.7, 2.8, 3.1, 3.3, 3.4])

diet_B = np.array([2.0, 2.5, 2.8, 2.2, 1.9, 2.4, 2.6, 2.1, 2.9, 2.3,
                   2.5, 2.7, 2.4, 2.3, 2.0, 2.2, 2.6, 2.1, 2.4, 2.3,
                   2.2, 2.4, 2.3, 2.1, 2.6, 2.2, 2.3, 2.4, 2.7, 2.8,
                   2.5, 2.3, 2.1, 2.4, 2.7, 2.9, 2.6, 2.5, 2.2, 2.3,
                   2.0, 2.6, 2.4, 2.3, 2.1, 2.7, 2.5, 2.8, 2.9, 2.2])

diet_C = np.array([3.5, 3.8, 3.9, 3.4, 3.0, 3.2, 3.1, 3.6, 3.7, 3.2,
                   3.4, 3.3, 3.6, 3.8, 3.5, 3.7, 3.9, 3.6, 3.3, 3.5,
                   3.4, 3.2, 3.1, 3.6, 3.7, 3.4, 3.3, 3.2, 3.9, 3.8,
                   3.5, 3.4, 3.3, 3.7, 3.9, 3.8, 3.6, 3.5, 3.4, 3.2,
                   3.0, 3.6, 3.3, 3.4, 3.8, 3.2, 3.5, 3.7, 3.9, 3.6])

# Combine data from all diets
all_data = np.concatenate((diet_A, diet_B, diet_C))

# Group labels
groups = ['A'] * len(diet_A) + ['B'] * len(diet_B) + ['C'] * len(diet_C)

# Calculate the one-way ANOVA
f_statistic, p_value = stats.f_oneway(diet_A, diet_B, diet_C)

# Print the results
print("F-Statistic:", f_statistic)
print("p-value:", p_value)

# Interpret the results
if p_value < 0.05:
    print("The p-value is less than 0.05, indicating a statistically significant difference among the diet groups.")
    print("There is evidence to suggest that at least one diet has a different mean weight loss compared to the others.")
else:
    print("The p-value is not significant at the 0.05 level, indicating no strong evidence of a difference among the diet groups.")
```

```python
#Question.10 A company wants to know if there are any significant differences in the average time it takes to
#complete a task using three different software programs: Program A, Program B, and Program C. They
#randomly assign 30 employees to one of the programs and record the time it takes each employee to
#complete the task. Conduct a two-way ANOVA using Python to determine if there are any main effects or
#interaction effects between the software programs and employee experience level (novice vs.
#experienced). Report the F-statistics and p-values, and interpret the results.
#Answer.10 : 
import numpy as np
import pandas as pd
import scipy.stats as stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Example data for time taken to complete a task
data = {
    'Software': np.random.choice(['A', 'B', 'C'], size=90),
    'Experience': np.random.choice(['Novice', 'Experienced'], size=90),
    'Time': np.random.normal(loc=20, scale=5, size=90)  # Mean=20, Std=5
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Perform two-way ANOVA
model = ols('Time ~ Software + Experience + Software:Experience', data=df).fit()
anova_table = anova_lm(model)

# Print the ANOVA table
print(anova_table)

# Interpret the results
software_p_value = anova_table.loc['Software', 'PR(>F)']
experience_p_value = anova_table.loc['Experience', 'PR(>F)']
interaction_p_value = anova_table.loc['Software:Experience', 'PR(>F)']

print("\nInterpretation:")
if software_p_value < 0.05:
    print("There is a significant main effect of software programs on the time to complete the task.")
else:
    print("There is no significant main effect of software programs.")

if experience_p_value < 0.05:
    print("There is a significant main effect of employee experience level on the time to complete the task.")
else:
    print("There is no significant main effect of employee experience level.")

if interaction_p_value < 0.05:
    print("There is a significant interaction effect between software programs and employee experience level.")
else:
    print("There is no significant interaction effect between software programs and employee experience level.")
```

```python
#Question.11 An educational researcher is interested in whether a new teaching method improves student test
#scores. They randomly assign 100 students to either the control group (traditional teaching method) or the
#experimental group (new teaching method) and administer a test at the end of the semester. Conduct a
#two-sample t-test using Python to determine if there are any significant differences in test scores
#between the two groups. If the results are significant, follow up with a post-hoc test to determine which
#group(s) differ significantly from each other.
#Answer.11 : 
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.stats.multicomp as multi

# Example data for test scores
control_group_scores = np.array([78, 85, 92, 73, 79, 88, 90, 82, 75, 81, 86, 88, 82, 84, 79, 91, 89, 85, 76, 80,
                                 84, 88, 86, 83, 80, 87, 81, 79, 75, 78, 81, 83, 82, 77, 89, 86, 90, 82, 79, 85,
                                 87, 84, 80, 83, 79, 86, 81, 89, 87, 85, 82])

experimental_group_scores = np.array([92, 88, 95, 87, 90, 85, 78, 94, 85, 82, 88, 83, 89, 92, 85, 90, 84, 87, 80, 84,
                                      86, 88, 91, 83, 86, 89, 84, 82, 90, 85, 87, 86, 84, 81, 89, 86, 83, 92, 88,
                                      85, 82, 87, 91, 84, 82, 79, 86, 90, 83, 85, 88, 92, 87])

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(control_group_scores, experimental_group_scores)

# Print the t-test results
print("Two-sample t-test:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Perform post-hoc test (Tukey's HSD)
data = np.concatenate((control_group_scores, experimental_group_scores))
group_labels = ['Control'] * len(control_group_scores) + ['Experimental'] * len(experimental_group_scores)
posthoc = multi.pairwise_tukeyhsd(data, group_labels, alpha=0.05)

# Print the post-hoc test results
print("\nPost-hoc test (Tukey's HSD):")
print(posthoc)
```

```python
#Question.12 A researcher wants to know if there are any significant differences in the average daily sales of three
#retail stores: Store A, Store B, and Store C. They randomly select 30 days and record the sales for each store
#on those days. Conduct a repeated measures ANOVA using Python to determine if there are anysignificant differences in 
#sales between the three stores. If the results are significant, follow up with a post-hoc test to determine which 
#store(s) differ significantly from each other.
#Answer.12 : 
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.stats.multicomp as multi

# Example data for daily sales
days = 30
store_A_sales = np.random.randint(50, 300, days)  # Random sales between 50 and 300
store_B_sales = np.random.randint(50, 300, days)
store_C_sales = np.random.randint(50, 300, days)

# Combine data from all stores
all_sales = np.concatenate((store_A_sales, store_B_sales, store_C_sales))

# Group labels
groups = ['Store A'] * days + ['Store B'] * days + ['Store C'] * days

# Create a DataFrame from the data
data = pd.DataFrame({'Sales': all_sales, 'Store': groups})

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(store_A_sales, store_B_sales, store_C_sales)

# Print the ANOVA results
print("One-way ANOVA:")
print("F-Statistic:", f_statistic)
print("p-value:", p_value)

# Perform post-hoc test (Tukey's HSD)
posthoc = multi.pairwise_tukeyhsd(data['Sales'], data['Store'], alpha=0.05)

# Print the post-hoc test results
print("\nPost-hoc test (Tukey's HSD):")
print(posthoc)
```
