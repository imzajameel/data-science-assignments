```python
#Go to this given URL and solve the following questions
#URL: https://www.youtube.com/@PW-Foundation/videos

#Question.1 : Write a python program to extract the video URL of the first five videos.
#Answer.1 :  
import requests
from bs4 import BeautifulSoup

# Replace this with the URL of the YouTube channel's videos page.
url = "https://www.youtube.com/@PW-Foundation/videos"

# Send an HTTP request to the URL and get the page's content.
response = requests.get(url)

# Check if the request was successful.
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup.
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the elements that contain video links.
    video_elements = soup.find_all("a", class_="yt-simple-endpoint style-scope ytd-grid-video-renderer")

    # Extract the video URLs of the first five videos.
    for i, video_element in enumerate(video_elements[:5]):
        video_url = "https://www.youtube.com" + video_element["href"]
        print(f"Video {i+1}: {video_url}")

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
#Question.5 : Write a python program to extract the time of posting of video for the first five videos.
#Answer.5 : 
import requests
from bs4 import BeautifulSoup

# Replace this with the URL of the YouTube channel's videos page.
url = "https://www.youtube.com/@PW-Foundation/videos"

# Send an HTTP request to the URL and get the page's content.
response = requests.get(url)

# Check if the request was successful.
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup.
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the elements that contain video posting times.
    time_elements = soup.find_all("div", class_="style-scope ytd-grid-video-renderer")

    # Extract the posting times of the first five videos.
    for i, time_element in enumerate(time_elements[:5]):
        time = time_element.find("a", class_="yt-simple-endpoint style-scope yt-formatted-string").text
        print(f"Video {i+1} Posting Time: {time}")

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
#Note :  Save all the data scraped in the above questions in a CSV file.
import requests
from bs4 import BeautifulSoup
import csv

# Replace this with the URL of the YouTube channel's videos page.
url = "https://www.youtube.com/@PW-Foundation/videos"

# Send an HTTP request to the URL and get the page's content.
response = requests.get(url)

# Check if the request was successful.
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup.
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the elements that contain video information.
    video_elements = soup.find_all("ytd-grid-video-renderer")

    # Extract the data of the first five videos.
    data = []
    for i, video_element in enumerate(video_elements[:5]):
        title = video_element.find("h3").text.strip()
        view_count = video_element.find("span", class_="style-scope ytd-grid-video-renderer").text.strip()
        posting_time = video_element.find("div", class_="style-scope ytd-grid-video-renderer").text.strip()
        data.append([title, view_count, posting_time])

    # Save the data to a CSV file.
    with open("youtube_data.csv", "w", newline="") as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(["Title", "View Count", "Posting Time"])
        csv_writer.writerows(data)

    print("Data saved to youtube_data.csv")

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)


```
