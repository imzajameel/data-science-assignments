```python
#Statistics advance 1 , assignment 1 
#Question.1 What is the Probability density function?
#Answer.1 A Probability Density Function (PDF) is a fundamental concept
#in probability and statistics. It describes the likelihood of a continuous
#random variable taking on a particular value or falling within a certain range
#of values. Unlike discrete random variables, which can take on only specific
#values (like the outcomes of rolling a die), continuous random variables can
#take on any value within a range (like the height of a person).

#mathematically, for a continuous random variable X, the probability density 
#function f(x) is defined such that the probability of X falling within a certain
#interval [a, b] is given by the integral of the PDF over that interval:

#(a ≤ X ≤ b) = ∫[a, b] f(x) dx

# PDF must satisfy two properties:

#non-negativity: f(x) ≥ 0 for all x.
#total area under the curve equals 1: ∫(-∞, ∞) f(x) dx = 1.

#In other words, the PDF provides a way to represent the distribution of probabilities for a
#continuous random variable across its entire range.

#For example, consider a normal distribution. The bell-shaped curve of a normal distribution 
#is described by its PDF, and different values of the parameters (mean and standard deviation) 
#will result in different PDFs, representing different probabilities of observing various values.

#It's important to note that while the PDF gives the relative likelihood of different outcomes, 
#it does not directly give the probability of a single point. The probability of a continuous
#random variable taking on any specific value is actually zero, because the number of possible
#values is infinite. Instead, we focus on the probabilities of intervals.

#In summary, a Probability Density Function is a mathematical function that characterizes the
#distribution of probabilities for a continuous random variable, helping us understand the
#likelihood of various outcomes or ranges of outcomes.
```

```python
#Question.2 What are the types of Probability distribution?
#Answer.2 There are several types of probability distributions, each
#dscribing the likelihood of different kinds of random variables and 
#their possibles. These distributions are essential in probability 
#theory and statistics for modeling and analyzing various real-world phenomena. Here
#are some of the most common types of probability distributions:

#Discrete Distributions:

#Bernoulli Distribution: Models a binary outcome (success/failure) with a single 
#parameter representing the probability of success.
#Binomial Distribution: Describes the number of successes in a fixed number of
#independent Bernoulli trials.
#Poisson Distribution: Models the number of events occurring in a fixed interval 
#of time or space, assuming a rare occurrence and constant rate.
#Geometric Distribution: Represents the number of trials needed for the first 
#success in a sequence of independent Bernoulli trials.

#Continuous Distributions:

#Uniform Distribution: Equally likely outcomes over a specified interval, often depicted as a flat
#horizontal line.
#Normal (Gaussian) Distribution: Characterized by its bell-shaped curve, this is one of the
#most common distributions in nature due to the Central Limit Theorem.
#Exponential Distribution: Models the time between events in a Poisson process, such as the
#time between phone calls at a call center.
#Gamma Distribution: Generalizes the exponential distribution and is often used to
#model waiting times and lifetimes.
#Beta Distribution: Describes random variables that have values within a fixed interval, 
#commonly used in Bayesian statistics.
#Chi-Squared Distribution: Arises in the context of the sum of squares of independent
#standard normal random variables, commonly used in hypothesis testing and confidence intervals.
#Cauchy Distribution: Known for its heavy tails, making it sensitive to outliers and 
#commonly used in robust statistics.
#Log-Normal Distribution: Results from exponentiating a normally distributed random
#variable and is often used to model quantities that are always positive (e.g., stock prices).

#Multivariate Distributions:

#Multinomial Distribution: Generalizes the binomial distribution to more than two outcomes.
#Multivariate Normal Distribution: Generalizes the normal distribution to multiple dimensions,
#commonly used in multivariate statistics.

#Other Distributions:

#Student's t-Distribution: Used for making inferences about the population mean when the sample
#size is small or the population standard deviation is unknown.
#F-Distribution: Used in ANOVA and other statistical tests to compare variances of multiple groups.
#Weibull Distribution: Commonly used to model lifetimes of objects, including reliability analysis.
#These are just a few examples of the many probability distributions that exist. Each distribution
#serves a specific purpose and is used to model different types of random variables and phenomena
#in various fields, including science, engineering, finance, and social sciences.
```

```python
#Question.3 Write a Python function to calculate the probability density function of a normal
#istribution with given and standard deviation at a given point.
#Answer.3 : 
import scipy.stats as stats

def normal_pdf(mean, std_dev, x):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution
    at a given point x, with the given mean and standard deviation.
    
    Args:
        mean (float): Mean of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.
        x (float): Point at which to calculate the PDF.
    
    Returns:
        float: Probability Density Function value at the given point x.
    """
    pdf_value = stats.norm.pdf(x, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean = 0
std_dev = 1
x_value = 1.5
pdf_at_x = normal_pdf(mean, std_dev, x_value)
print(f"PDF at x = {x_value}: {pdf_at_x}")
```

```python
#Question.4 What are the properties of Binomial distribution? Give two examples of events where binomial
#distribution can be applied.
#Answer.4 he binomial distribution is a fundamental probability distribution that describes
#the number of successes in a fixed number of independent Bernoulli trials (binary outcomes:
#success or failure) with the same probability of success in each trial. The properties
#of the binomial distribution include:

#Parameters:

#n: Number of trials.
#p: Probability of success on each trial.
#Probability Mass Function (PMF):
#The probability of getting exactly k successes in n trials is given by:
#P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)
#where "n choose k" represents the binomial coefficient.

#Mean and Variance:
    
#Mean (μ) = n * p
#Variance (σ^2) = n * p * (1 - p)

#Shape:
#The shape of the binomial distribution is influenced by n (number of trials) and p 
#(probability of success). As n increases while keeping p fixed, the distribution becomes 
#more symmetric and approaches a normal distribution.

#Use of the Cumulative Distribution Function (CDF):
#The cumulative distribution function gives the probability that the number of
#successes is less than or equal to a certain value.

#Two examples of events where the binomial distribution can be applied are:

#Coin Flipping:
#Suppose you flip a fair coin (with a 50% chance of heads and 50% chance of tails)
#10 times. You can model the number of times you get heads as a binomial distribution 
#with parameters n = 10 and p = 0.5.

#Quality Control:
    
#A manufacturing process produces items, and each item can be either defective (failure)
#or non-defective (success). Let's say you're interested in the number of defective items 
#in a sample of 50 items produced by this process. You could model this using a binomial
#distribution with parameters n = 50 and a given probability p of an item being defective
#based on historical data.

#In both examples, the binomial distribution helps you analyze and predict the
#likelihood of getting a certain number of successes (e.g., heads, defective items)
#in a fixed number of trials or observations.
```

```python
#Question.5 Generate a random sample of size 1000 from a binomial distribution with probability of
#success 0.4 and plot a histogram of the results using matplotlib.
#Answer.5 : 
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1000  # Sample size
p = 0.4   # Probability of success

# Generate random sample from binomial distribution
random_sample = np.random.binomial(n, p, size=n)

# Plot histogram
plt.hist(random_sample, bins=np.arange(-0.5, n + 1.5, 1), align='mid', density=True, alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Histogram of Binomial Distribution')
plt.grid(True)
plt.show()
```

```python
#Question.6 Write a Python function to calculate the cumulative distribution function of a 
#Poisson distribution with given mean at a given point.
#Answer.6 : 
import scipy.stats as stats

def poisson_cdf(mean, x):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution
    at a given point x, with the given mean.
    
    Args:
        mean (float): Mean of the Poisson distribution.
        x (int): Point at which to calculate the CDF.
    
    Returns:
        float: Cumulative Distribution Function value at the given point x.
    """
    cdf_value = stats.poisson.cdf(x, mu=mean)
    return cdf_value

# Example usage
mean = 3
x_value = 2
cdf_at_x = poisson_cdf(mean, x_value)
print(f"CDF at x = {x_value}: {cdf_at_x}")
```

```python
#Question.7 How Binomial distribution different from Poisson distribution?
#Answer.7 The Binomial and Poisson distributions are two distinct probability
#distributions used to model different types of random processes. While they
#share some similarities, they are applied in different contexts and have
#distinct characteristics. 

#Here's how they differ:

#number of Parameters: The Binomial distribution has two parameters (n and p),
#while the Poisson distribution has one parameter (λ).

#application: Binomial distribution is used when dealing with a fixed number 
#of trials with binary outcomes, while Poisson distribution is used for modeling rare
#and random events in a fixed interval.

#Probability of Success: In the Binomial distribution, the probability of success
#remains constant across all trials. In the Poisson distribution, the parameter λ 
#represents the average rate of event occurrences.

#Events: The Binomial distribution deals with discrete events involving successes
#and failures. The Poisson distribution deals with the occurrence of rare events
#in continuous time or space.

#Relation: In certain cases, the Poisson distribution can be considered a limiting 
#case of the Binomial distribution when the number of trials is very large and the 
#probability of success is very small, while keeping the expected number of successes 
#constant (λ = n * p).
```

```python
#Question.8 Generate a random sample of size 1000 from a Poisson distribution with mean 5 
#and calculate the sample mean and variance.
#Answer.8 : 

import numpy as np

# Parameters
sample_size = 1000
mean = 5

# Generate random sample from Poisson distribution
random_sample = np.random.poisson(mean, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

```python
#Question.9 How mean and variance are related in Binomial distribution and Poisson distribution?
#Answer.9 The mean and variance of both the Binomial and Poisson distributions are related
#to the distribution's parameters and characteristics, but they are related differently 
#for each distribution.

#Binomial Distribution:

#For a Binomial distribution with parameters n (number of trials) and p 
#(probability of success on each trial), the mean and variance are given by:

#Mean (μ) = n * p
#Variance (σ^2) = n * p * (1 - p)

#In the Binomial distribution, the variance is influenced by both the number of
#trials n and the probability of success p. As n increases while keeping p fixed,
#the variance tends to increase, indicating greater variability in outcomes. Similarly,
#as p approaches 0 or 1 while keeping n fixed, the variance also tends to decrease.

#Poisson Distribution:
#For a Poisson distribution with parameter λ (average rate of event occurrences),
#the mean and variance are both equal to λ:

#Mean (μ) = λ
#Variance (σ^2) = λ

#In the Poisson distribution, the mean and variance are directly related and are
#both equal to the parameter λ. This relationship implies that the spread or 
#variability of the distribution is solely determined by the average rate of
#event occurrences. As λ increases, both the mean and variance increase, indicating 
#a broader spread of events.

#Comparison:
#In summary, in the Binomial distribution, the variance is influenced by both the 
#number of trials and the probability of success, while in the Poisson distribution, 
#the variance is solely determined by the mean (average rate of occurrences). The 
#relationship between mean and variance is different for each distribution, but
#both provide insights into the distribution's characteristics and the spread of 
#possible outcomes.
```

```python
#Question.10 In normal distribution with respect to mean position, where does the
#least frequent data appear?
#Answer.10 In a normal distribution, which is symmetric and bell-shaped, the least
#frequent data points occur in the tails of the distribution, farthest from the mean. 
#The tails of the normal distribution represent the extreme values, and these values 
#occur with lower frequencies compared to the values closer to the mean.

#The mean of a normal distribution is the central point of the distribution and 
#represents the highest point of the curve. As you move away from the mean in 
#either direction towards the tails, the frequency of data points decreases. 
#This is a key characteristic of the normal distribution's symmetric and bell-shaped
#nature.

#In statistical terms, the "tails" of a normal distribution refer to the portions of the 
#distribution that are located beyond a certain number of standard deviations from
#the mean. The data points in these tails are the least frequent and are considered 
#outliers or extreme values. The percentage of data within a certain number of standard
#deviations from the mean is given by the empirical rule (also known as the 68-95-99.7 
#rule), which states that approximately:

#About 68% of the data falls within one standard deviation of the mean.
#About 95% falls within two standard deviations.
#About 99.7% falls within three standard deviations.
#So, the data points that are more than three standard deviations away from the
#mean (in the tails) are the least frequent and most extreme values in a normal distribution.
```
