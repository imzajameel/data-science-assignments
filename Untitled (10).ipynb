```python
#Statistics_Advance_Assignment_4
#Question.1 What is the difference between a t-test and a z-test? Provide an example scenario where you would
#use each type of test.
#Answer.1 The t-test and z-test are both statistical hypothesis tests used to make inferences about population 
#parameters based on sample data. However, they have differences in terms of assumptions and application scenarios.
#The main distinction between the two lies in the information available about the population standard deviation.

#Z-Test:
#The z-test is used when the population standard deviation is known or when the sample size is large (typically n≥30).
#It is based on the standard normal distribution (Z-distribution). The formula for the z-test statistic is similar
#to that of the t-test, but it uses the population standard deviation instead of the sample standard deviation.

#Example Scenario for Z-Test:
    
#Suppose a company wants to test whether a new manufacturing process has led to a significant increase in the
#average weight of their products. They have access to historical data on the population standard deviation 
#of product weights. In this case, a z-test would be appropriate because the population standard deviation is known.

#T-Test : 
#The t-test is used when the population standard deviation is unknown or when the sample size is small (typically  n<30). 
#It is based on the t-distribution, which has heavier tails compared to the standard normal distribution. 
#The t-test uses the sample standard deviation and has slightly different formulae based on the specific 
#scenario (one-sample, two-sample, paired-sample).

#Example Scenario for T-Test:
    
#Consider a medical study comparing the effectiveness of two different drugs in reducing blood pressure. 
#Two groups of patients are given different drugs, and their blood pressure measurements are recorded. Since 
#the population standard deviation of blood pressure is unknown, a t-test would be appropriate to compare the
#means of the two groups.

#In summary:

#Use a z-test when the population standard deviation is known or when the sample size is large.
#Use a t-test when the population standard deviation is unknown or when the sample size is small.
#It's important to choose the appropriate test based on the characteristics of your data and the assumptions of the tests.
#Both tests are used for hypothesis testing and making inferences about population parameters based on sample data.
```

```python
#Question.2 Differentiate between one-tailed and two-tailed tests.
#Answer.2 One-tailed and two-tailed tests are concepts in hypothesis testing that refer to the 
#directionality of the statistical hypothesis and the associated critical region. They determine 
#whether the test focuses on a specific direction of difference or change, or if it is concerned with
#deviations in both directions. Let's differentiate between one-tailed and two-tailed tests:

#One-Tailed Test:
#In a one-tailed test, the null hypothesis is tested against an alternative hypothesis 
#that specifies a particular direction of difference or change. The critical region is located entirely
#on one side of the distribution, corresponding to the chosen direction.

#There are two types of one-tailed tests:

#Upper-tailed Test: The critical region is in the upper tail of the distribution. It is used to 
#test if the sample statistic is significantly greater than a specified value.
#Lower-tailed Test: The critical region is in the lower tail of the distribution. It is used 
#to test if the sample statistic is significantly smaller than a specified value.

#Example of One-Tailed Test:
    
#Suppose a company claims that their new product increases customer satisfaction by at least 10%. A one-tailed 
#test would be used to determine if the sample data provides sufficient evidence to support this claim. The
#null hypothesis might state that the increase is less than or equal to 10%, while the alternative hypothesis
#would state that the increase is greater than 10%.

#Two-Tailed Test:
#In a two-tailed test, the null hypothesis is tested against an alternative hypothesis that allows
#for deviations in both directions. The critical region is divided between the two tails of the distribution,
#representing extreme values in both directions.

#A two-tailed test is used when you want to test if a sample statistic is significantly different from a 
#specified value, without specifying whether it is greater or smaller.

#Example of Two-Tailed Test:
#Suppose a manufacturer claims that their product has a mean weight of 10 pounds. A two-tailed test would 
#be used to determine if the sample data provides enough evidence to suggest that the mean weight is significantly 
#different from 10 pounds, without specifying whether it is greater or smaller.

#In summary:

#One-tailed tests are used when you have a specific direction in mind (increase, decrease) and want to test that direction.
#Two-tailed tests are used when you want to test for differences in both directions without specifying a specific direction.
#The choice between one-tailed and two-tailed tests depends on the research question, the directionality of 
#the hypothesis, and the nature of the investigation.
```

```python
#Question.3 Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
#each type of error.
#Answer.3 n hypothesis testing, Type 1 and Type 2 errors are two potential mistakes that can occur when making
#decisions about a null hypothesis based on sample data. These errors are inherent to the process of hypothesis 
#testing and involve the incorrect acceptance or rejection of the null hypothesis.

#Type 1 Error (False Positive):
    
#A Type 1 error occurs when the null hypothesis is incorrectly rejected when it is actually true. In other words,
#the test incorrectly indicates a significant effect or difference that does not exist in the population.

#Symbolically, a Type 1 error is denoted as α (alpha), which represents the significance level of the test. It 
#is the probability of rejecting the null hypothesis when it is true.

#Example Scenario of Type 1 Error:
    
#Suppose a medical researcher is testing a new drug to determine if it reduces the incidence of a specific
#condition. The null hypothesis states that the drug has no effect. If the researcher incorrectly rejects the
#null hypothesis and concludes that the drug is effective when it actually isn't, this would be a Type 1 error.

#Type 2 Error (False Negative):
    
#A Type 2 error occurs when the null hypothesis is incorrectly accepted when it is actually false. In other words, 
#the test fails to detect a significant effect or difference that does exist in the population.

#Symbolically, a Type 2 error is denoted as β (beta), which represents the probability of failing to reject 
#the null hypothesis when it is false. The complementary probability (1−β) is known as the power of the test, which 
#represents the probability of correctly rejecting the null hypothesis when it is false.

#Example Scenario of Type 2 Error:
    
#Continuing with the previous example, if the researcher fails to reject the null hypothesis and concludes
#that the drug has no effect, even though it actually does reduce the condition's incidence, this would be a Type 2 error.

#In summary:

#Type 1 Error (False Positive): Rejecting the null hypothesis when it is true.
#Type 2 Error (False Negative): Failing to reject the null hypothesis when it is false.
#The balance between Type 1 and Type 2 errors is a fundamental consideration in hypothesis testing. As the significance 
#level (α) decreases, the risk of Type 1 errors decreases, but the risk of Type 2 errors may increase. Researchers 
#often aim to find a balance that minimizes both types of errors based on the specific context and consequences of 
#the decision.
```

```python
#Question.4 Explain Bayes's theorem with an example.
#Answer.4 Bayes's Theorem is a fundamental concept in probability theory and statistics that describes
#how to update the probability of a hypothesis (an event) based on new evidence or information. It helps
#us revise our beliefs about the likelihood of an event occurring when new data becomes available.
#The theorem is named after Thomas Bayes, an 18th-century mathematician.

#Bayes's Theorem is expressed as:

#P(A∣B)= P(B∣A)⋅P(A) / P(B)

#Where:

#P(A∣B) is the probability of event A occurring given that event B has occurred.

#P(B∣A) is the probability of event B occurring given that event A has occurred.

#P(A) is the prior probability of event A.

#P(B) is the prior probability of event B.

#Example of Bayes's Theorem: Medical Diagnosis

#Let's consider a medical diagnosis scenario to illustrate Bayes's Theorem:

#Suppose there's a rare disease called Disease X that affects 1% of the population. A medical test has
#been developed to detect this disease. The test is not perfect; it has a false positive rate of 5% 
#(probability of testing positive given that the person does not have the disease) and a false negative 
#rate of 2% (probability of testing negative given that the person has the disease).

#We want to calculate the probability that a person actually has Disease X given that they tested positive
#(posterior probability).

#Let's define the events:

#Event A: Having Disease X
#Event B: Testing Positive

#We are given:

#P(A)=0.01 (1% of the population has the disease)

#P(B∣A)=0.98 (98% accuracy of testing positive if you have the disease)

#P(¬B∣¬A)=0.95 (95% accuracy of testing negative if you don't have the disease)
               
#We need to find:
#P(A∣B)

#Using Bayes's Theorem:

#P(A∣B)= P(B∣A)⋅P(A)/P(B)

#We know that :

#P(B)=P(B∣A)⋅P(A)+P(B∣¬A)⋅P(¬A), where P(¬A)=1−P(A) is the probability of not having the disease.

#Plugging in the values: P(A∣B) = 0.98⋅0.01 / (0.98⋅0.01+0.05⋅0.99 ) ≈ 0.98⋅0.01
               
#So, even if a person tests positive, the probability that they actually have Disease X is only around 16.4%, due to the
#low prevalence of the disease and the imperfect accuracy of the test.

#Bayes's Theorem demonstrates how prior probabilities and conditional probabilities can be combined to update our beliefs 
#based on new evidence.               
```

```python
#Question.5 What is a confidence interval? How to calculate the confidence interval, explain with an example.
#Answer.5 A confidence interval is a range of values that provides an estimated range of a population parameter,
#along with a level of confidence that the true parameter lies within that range. It is a way to quantify the
#uncertainty or variability associated with a sample estimate.

#In other words, a confidence interval gives you a sense of how confident you can be that the true population 
#parameter (e.g., mean, proportion) falls within a certain range of values based on the sample data.

#Calculating a Confidence Interval:

#The general formula for calculating a confidence interval for a population parameter is:

#Confidence Interval = Point Estimate ± Margin of Error

#Where:

#Point Estimate: The sample statistic (e.g., sample mean, sample proportion) that estimates the population parameter.
#Margin of Error: The range around the point estimate that accounts for the variability of the sample and provides
#a measure of uncertainty.

#To calculate the margin of error, you typically need the following information:

#The chosen level of confidence (e.g., 95%, 99%).
#The standard error of the sample statistic, which is a measure of how much the sample statistic varies from sample to sample.
#Example of Calculating a Confidence Interval:

#Suppose you are conducting a survey to estimate the average income of people in a certain city. You collect a random 
#sample of 100 individuals and find that the sample mean income is $50,000, and the sample standard deviation is $8,000.

#You want to calculate a 95% confidence interval for the true average income of the population.

#Step 1: Calculate the Margin of Error
#The margin of error depends on the critical value from the chosen confidence level and the standard error of the 

#sample mean (SE):

#Margin of Error = Critical Value × SE

#Step 2: Calculate the Standard Error
#The standard error of the sample mean (SE) is calculated using the formula:
    
#SE = Sample Standard Deviation/Sample Size

#Step 3: Find the Critical Value
#For a 95% confidence interval, the critical value from the standard normal distribution is approximately 
#1.96 (you can also use a t-distribution critical value depending on sample size and distribution assumptions).

#Step 4: Calculate the Confidence Interval
#Confidence Interval = Point Estimate ± Margin of Error
#Confidence Interval = Point Estimate±Margin of Error

#Plugging in the values:
    
#Confidence Interval = $50,000 ± $1,568
                   # = $48,432 to $51,568


#This means that you can be 95% confident that the true average income of the population falls within the range
#of $48,432 to $51,568 based on the sample data.
```

```python
#Question.8 What is the margin of error in a confidence interval? How does sample size affect the margin of error?
#Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
#Answer.8 The margin of error in a confidence interval is a measure of the uncertainty or variability associated with
#the estimation of a population parameter based on a sample. It represents the range around the point estimate
#(sample statistic) within which the true population parameter is likely to fall with a certain level of confidence.

#In other words, the margin of error quantifies how much the sample estimate might differ from the true population value.


#The formula for calculating the margin of error in a confidence interval is:

#Margin of Error = Critical Value ×  Sample Size

#Where:

#Critical Value: This value is determined by the chosen level of confidence and the distribution (e.g., normal
#distribution, t-distribution).

#Sample Standard Deviation: A measure of the variability of the sample data.
#Sample Size: The number of observations in the sample.

#Impact of Sample Size on Margin of Error:

#A larger sample size generally leads to a smaller margin of error. When the sample size increases, the denominator 
#of the margin of error formula (sample_size^1/2)
#becomes larger, resulting in a smaller overall margin of error. In other words, larger sample sizes provide more
#information and reduce the uncertainty associated with the estimation.

#Example Scenario:

#Suppose a company wants to estimate the average time it takes for customers to complete an online survey.
#They collect two different samples:

#Sample 1: 

#n=100 (larger sample size)

#Sample 2: 

#n=25 (smaller sample size)

#For both samples, the point estimate (sample mean) is found to be 8 minutes, and the sample standard deviation
#is 2 minutes.

#Using a 95% confidence level and a critical value of approximately 1.96 (for a normal distribution):

#For Sample 1:
#Margin of Error = 1.96 × (2 / (100 ^1/2)) = 0.392 minutes

#For Sample 2:
#Margin of Error = 1.96 × (2/(25^1/2))  = 0.784 minutes

#As you can see, the margin of error for Sample 1 (larger sample size) is smaller than the margin of error for
#Sample 2 (smaller sample size). This means that the confidence interval for the true population mean is narrower
#for Sample 1, indicating a more precise estimate of the parameter.

#In summary, larger sample sizes result in smaller margins of error and narrower confidence intervals,
#leading to more accurate and precise estimates of population parameters.
```

```python
#Question.9 Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
#standard deviation of 5. Interpret the results.
#Answer.9 The z-score (also known as the standard score) measures how many standard deviations a data point
#is away from the mean of a population. It indicates how relatively unusual or extreme a particular data point
#is within the context of the population's distribution.

#The formula to calculate the z-score is :

#z =  (x−μ )/ σ

#Where:

#z is the z-score.

#x is the value of the data point.

#μ is the population mean.

#σ is the population standard deviation.

#Substitute the values into the formula:

#z = (75-70)/5 = 1

#Interpreting the Results:
    
#A z-score of 1 means that the data point (75) is 1 standard deviation above the population mean (70). 
#In other words, the data point is relatively higher than the average value in the population by one standard deviation.

#Z-scores are used to compare data points across different distributions and to determine how far a particular 
#observation deviates from the mean. Positive z-scores indicate values above the mean, while negative z-scores
#indicate values below the mean. The magnitude of the z-score indicates how far the data point is from the mean 
#in terms of standard deviations.
```

```python
#Question.10 In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
#of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
#significantly effective at a 95% confidence level using a t-test.
#Answer.10  To conduct a hypothesis test to determine if the new weight loss drug is significantly effective at a 95% 
#confidence level using a t-test, we need to set up the null and alternative hypotheses, calculate the test statistic, 
#and compare it to the critical value or p-value.

#Null Hypothesis :  The weight loss drug is not significantly effective (μ = 0, where μ is the population mean weight loss).

#Alternative Hypothesis : The weight loss drug is significantly effective (μ<0).

#Given the sample information:

#Sample Mean = 6 pounds
#Sample Standard Deviation  = 2.5 pounds
#Sample Size = 50

#Assumptions:

#The population of weight loss differences is approximately normally distributed.
#Steps to Conduct the Hypothesis Test:

#Calculate the t-test statistic:

#t =  (x−μ) / s / n^(1/2)

#Determine the degrees of freedom :

#df = n − 1

#Find the critical value or p-value associated with the t-test statistic and degrees of freedom.

#Compare the calculated t-test statistic to the critical value or p-value to make a decision.

#State the conclusion in terms of the original problem.

#Since the problem does not provide a specific value for the population mean (μ), we will assume a value 
#of 0 in this example.

#Let's calculate the t-test statistic and determine the conclusion:

#t = ( 6 - 0 ) / 2.5 / 50^(1/2) ≈ 17.677

#Degrees of Freedom: 

#df= 50 − 1 = 49

#Using a t-distribution table or calculator, we find that the critical value for a one-tailed test at a 95% 
#confidence level with 

#df=49 is approximately -1.676.

#Since the calculated t-test statistic (17.677) is much larger than the critical value (-1.676), we can 
#reject the null hypothesis. This means that there is sufficient evidence to conclude that the weight 
#loss drug is significantly effective at a 95% confidence level.

#In summary, based on the results of the t-test, the new weight loss drug is found to be significantly effective
#in helping participants lose weight.
```

```python
#Question.13 A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations
#has a mean of 65. Calculate the 90% confidence interval for the true population mean.
#Answer.13 To calculate the 90% confidence interval for the true population mean, we can use the formula for the 
#confidence interval of a population mean when the population standard deviation is known:

#Confidence Interval = Sample Mean ± (Critical Value) * (Population Standard Deviation / √Sample Size)

#Given:
#Population Mean (μ) = 60
#Population Standard Deviation (σ) = 8
#Sample Mean (x̄) = 65
#Sample Size (n) = 50

#Step 1: Find the Critical Value:
#For a 90% confidence interval and a symmetric distribution (like the normal distribution), we need to
#find the critical value corresponding to the middle 95% of the distribution. You can use a standard normal 
#distribution table or a calculator. The critical value (z) for a 90% confidence interval is approximately 1.645.

#Step 2: Calculate the Margin of Error:
#Margin of Error = (Critical Value) * (Population Standard Deviation / √Sample Size)

#Margin of Error = 1.645 * (8 / √50) ≈ 2.317

#Step 3: Calculate the Confidence Interval:
#Lower Limit = Sample Mean - Margin of Error
#Upper Limit = Sample Mean + Margin of Error

#Lower Limit = 65 - 2.317 ≈ 62.683
#Upper Limit = 65 + 2.317 ≈ 67.317

#Step 4: Write the Confidence Interval:
#The 90% confidence interval for the true population mean is approximately (62.683, 67.317).

#Therefore, we can be 90% confident that the true population mean lies between 62.683 and 67.317 based on 
#the given sample data.
```
